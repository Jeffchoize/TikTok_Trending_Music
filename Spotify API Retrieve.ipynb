```python
# Remove the duplicated rows
dedup_data = data.drop_duplicates()
dedup_data = dedup_data.reset_index(drop = True)
```

```python
# base URL of all Spotify API endpoints

BASE_URL = 'https://api.spotify.com/v1/'

followers = []
explicit = []


for i in range(3000,4645):
    
    # Track/Artist ID from the csv file
    artist_id = dedup_data["artist_id"][i]
    track_id = dedup_data["track_id"][i]

    # actual GET request with proper header
    arres = requests.get(BASE_URL + 'artists/' + artist_id, headers=headers)
    trackres = requests.get(BASE_URL + 'tracks/' + track_id, headers=headers)
    
    # load the json content
    res_artist = json.loads(arres.content)
    res_track = json.loads(trackres.content)
    
    # normalize json to dataframe
    pd_artist = pd.json_normalize(res_artist)
    pd_track = pd.json_normalize(res_track)
    
    # extract the followers of the artists and explicitness of the track
    followers.append(pd_artist['followers.total'][0])
    explicit.append(pd_track['explicit'][0])

dedup_data['followers']=followers
dedup_data['explicit']=explicit
```
